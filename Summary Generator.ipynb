```python
import os
import re
# Pastikan Anda telah menginstal pustaka PyMuPDF (MuPDF) dengan perintah: pip install pymupdf

# Tentukan path folder tempat PDF berada
folder_path = 'pdf_putusan_nadia'

# Fungsi untuk membaca nama file PDF di dalam folder
def read_pdf_filenames(folder_path):
    pdf_files = [file for file in os.listdir(folder_path) if file.endswith('.pdf')]
    return pdf_files

# Mendapatkan daftar nama file PDF di dalam folder
pdf_filenames = read_pdf_filenames(folder_path)

all_angka_list = []
# Menampilkan nama file PDF
for filename in pdf_filenames:
    print(filename)
       # Menggunakan ekspresi reguler untuk mengekstrak angka antara kata "putusan" dan "pid"
    match = re.search(r'putusan_(\d+)_pid', filename)
    
    # Jika ada kecocokan, ambil angka dan ambil 50 angka pertama
    if match:
        angka_str = match.group(1)
        
        # Menggabungkan seluruh angka menjadi satu angka utuh
        angka_utuh = int(angka_str)
        
        # Membuat array list untuk menampung seluruh 50 angka pertama
        angka_list = [angka_utuh]
        print(angka_utuh)
        # Menambahkan angka utuh ke list baru
        all_angka_list.append(angka_utuh)
        # print(angka_list)
    else:
        print("Tidak ada angka yang ditemukan antara kata 'putusan' dan 'pid' dalam nama file.")
print(all_angka_list)
```

```python
# Mengambil 50 item pertama dari list
angka_50_pertama = all_angka_list[:50]

# Menampilkan 50 item pertama
print(angka_50_pertama)
```
